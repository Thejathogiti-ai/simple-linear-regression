### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(20,5))
plt.title("Customer Grouping Dendogram")

Dend=shc.dendrogram(shc.linkage(samples,metric='euclidean',method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
df['Customer_grouping_H']=model.fit_predict(samples)
df.head()
```

### Comparision of Clusters

```python
import seaborn as sns 
sns.scatterplot(df['Customer_grouping'])
```

```python
sns.scatterplot(df['Customer_grouping_H'])
```

```python
from sklearn.metrics import silhouette_score
silhouette=silhouette_score(samples,df['Customer_grouping'])
print(silhouette)
```

```python
from sklearn.metrics import silhouette_score
silhouette=silhouette_score(samples,df['Customer_grouping_H'])
print(silhouette)
```

```python
k_values=[2,3,4,5,6,7,8]
silhouette_scores=[]
for k in k_values:
    model=KMeans(n_clusters=k,random_state=99,max_iter=100,n_init='auto')
    df['silpreds']=model.fit_predict(samples)
    score=silhouette_score(samples,df.silpreds)
    silhouette_scores.append(score)
    print(f"k= {k},silhouette score={score:.2f}")
```

```python
plt.plot(k_values,silhouette_scores,marker='o')
plt.xlabel("number of Clusters")
plt.ylabel("silhouette score")
plt.title("Number of Clusters(k) and silhouette_scores")
plt.show
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

visualizer=SilhouetteVisualizer(model,colors='yellowbrick')
visualizer.fit(samples)
visualizer.show()
```
