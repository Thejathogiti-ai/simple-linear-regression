```python
df['Admissionchance']=np.where(df['Chance of Admit ']>0.8,1,0)
df['Admissionchance']
```

```python
Df=df.drop(['Chance of Admit '],axis=1)
```

```python
Df.head()
```

```python
from sklearn.model_selection import train_test_split
x=Df.drop(['Admissionchance'],axis=1)
y=Df['Admissionchance']

X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=77)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
Admissionchance=['0','1']
```

```python
from sklearn import tree
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

independentvariables=list(x.columns)
Admissionchance=['1','0']

fig,axes=plt.subplots(nrows=1,ncols=1,dpi=300,figsize=(4,5))
plot_tree(dt,
         feature_names=list(x.columns),
         class_names=Admissionchance,
         filled=True,
         fontsize=2)
```

```python
Train=pd.concat([X_train,y_train],axis=1)
Train.head()
```

```python
Train['preds']=dt.predict(X_train)
Train.head()
```

```python
from sklearn.metrics import classification_report
report=classification_report(Train['preds'],Train['Admissionchance'])
print(report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

grid=GridSearchCV(DecisionTreeClassifier(random_state=99),
                params,
                verbose=1,
                cv=3)
grid.fit(X_train,y_train)
```

```python
DT=DecisionTreeClassifier(max_depth=3, min_samples_leaf=50, min_samples_split=200,
                       random_state=99)
DT.fit(X_train,y_train)
```

```python
Admissionchance=['0','1']
```

```python
from sklearn import tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig,axes=plt.subplots(nrows=1,ncols=1,dpi=300,figsize=(4,5))

plot_tree(DT,
         feature_names=list(x.columns),
         class_names=Admissionchance,
         filled=True,
         node_ids=True,
         fontsize=2)
```

```python
Train['predicted']=DT.predict(X_train)
Train.head()
```

```python
report=classification_report(Train['predicted'],Train['Admissionchance'])
print(report)
```

### Randomforest

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(max_depth=3,
                             min_samples_leaf=50,
                             min_samples_split=200,
                             n_estimators=20,
                       random_state=99,
                            max_features="sqrt")
model.fit(X_train,y_train)
```

```python
from sklearn.tree import export_graphviz
import pydot

tree=model.estimators_[5]
Admissionchance=['0','1']
export_graphviz(tree,
                out_file='abc.dot',
                filled=True,
                rounded=True,
                class_names=Admissionchance,
                feature_names=list(x.columns))

(graph,)=pydot.graph_from_dot_file('abc.dot')
graph.write_png('tree.png')

from IPython.display import Image
Image(filename='tree.png')
```

```python
Train['Forestpreds']=model.predict(X_train)
Train.head()
```

```python
report=classification_report(Train['Forestpreds'],Train['Admissionchance'])
print(report)
```

```python
Test['Forestpreds']=model.predict(X_test)
Test.head()
```

```python
report=classification_report(Test['Forestpreds'],Test['Admissionchance'])
print(report)
```
